```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.metrics import balanced_accuracy_score, confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
```

---

## Split the Data into Training and Testing Sets

### Step 1: Read the `lending_data.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
# YOUR CODE HERE!
credit_data = pd.read_csv('Resources/lending_data.csv')
# Review the DataFrame
# YOUR CODE HERE!
credit_data.head()
```

### Step 2: Create the labels set (`y`)  from the “loan_status” column, and then create the features (`X`) DataFrame from the remaining columns.

```python
# Separate the data into labels and features

# Separate the y variable, the labels
# YOUR CODE HERE!]
y = credit_data['loan_status'].copy()
# Separate the X variable, the features
# YOUR CODE HERE!
X = credit_data.drop(columns='loan_status').copy()
```

```python
# Review the y variable Series
# YOUR CODE HERE!
y.head()
```

```python
# Review the X variable DataFrame
# YOUR CODE HERE!
X.head()
```

### Step 3: Check the balance of the labels variable (`y`) by using the `value_counts` function.

```python
# Check the balance of our target values
# YOUR CODE HERE!
y.value_counts()
```

### Step 4: Split the data into training and testing datasets by using `train_test_split`.

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split

# Split the data using train_test_split
# Assign a random_state of 1 to the function
# YOUR CODE HERE!
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
X_train.shape
```

---

## Create a Logistic Regression Model with the Original Data

###  Step 1: Fit a logistic regression model by using the training data (`X_train` and `y_train`).

```python
# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
# YOUR CODE HERE!
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=200,
                                random_state=1)
classifier
# Fit the model using training data
# YOUR CODE HERE!
classifier.fit(X_train, y_train)
```

### Step 2: Save the predictions on the testing data labels by using the testing feature data (`X_test`) and the fitted model.

```python
# Make a prediction using the testing data
# YOUR CODE HERE!
predictions = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
results.head(10)
```

### Step 3: Evaluate the model’s performance by doing the following:

* Calculate the accuracy score of the model.

* Generate a confusion matrix.

* Print the classification report.

```python
# Print the balanced_accuracy score of the model
# YOUR CODE HERE!
accuracy_score(y_test, predictions)
```

```python
# Generate a confusion matrix for the model
# YOUR CODE HERE!
confusion_matrix(y_test, predictions)
```

```python
# Print the classification report for the model
# YOUR CODE HERE!


print(classification_report(y_test, predictions))
```

**Question:** How well does the logistic regression model predict both the `0` (healthy loan) and `1` (high-risk loan) labels?

**Answer:** The model is 99% accurate.

## Predict a Logistic Regression Model with Resampled Training Data

### Step 1: Use the `RandomOverSampler` module from the imbalanced-learn library to resample the data. Be sure to confirm that the labels have an equal number of data points. 

```python
# Import the RandomOverSampler module form imbalanced-learn
from imblearn.over_sampling import RandomOverSampler

# Instantiate the random oversampler model
# # Assign a random_state parameter of 1 to the model
# YOUR CODE HERE!
ros = RandomOverSampler(random_state=1)
# Fit the original training data to the random_oversampler model
# YOUR CODE HERE!
X_res, y_res = ros.fit_resample(X, y)
```

```python
# Count the distinct values of the resampled labels data
# YOUR CODE HERE!
X_res.nunique()
```

### Step 2: Use the `LogisticRegression` classifier and the resampled data to fit the model and make predictions.

```python
# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
# YOUR CODE HERE!
log_model = LogisticRegression(solver='lbfgs', random_state = 1)
# Fit the model using the resampled training data
# YOUR CODE HERE!
log_model.fit(X_res, y_res)
# Make a prediction using the testing data
# YOUR CODE HERE!
prediction = log_model.predict(X_res)
```

### Step 3: Evaluate the model’s performance by doing the following:

* Calculate the accuracy score of the model.

* Generate a confusion matrix.

* Print the classification report.

```python
# Print the balanced_accuracy score of the model 
# YOUR CODE HERE!
resampled_ac = balanced_accuracy_score(y_res, prediction)
print(resampled_ac)
```

```python
# Generate a confusion matrix for the model
# YOUR CODE HERE!
reseampled_train_matrix = confusion_matrix(y_res, prediction)
```

```python
# Print the classification report for the model
# YOUR CODE HERE!.
resampled_train_report = classification_report(y_res, prediction)
print(resampled_train_report)
```
